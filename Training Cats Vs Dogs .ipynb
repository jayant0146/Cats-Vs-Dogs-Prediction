## Import Modules

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

## Data Gathering 

```python
path = r'C:\Users\ACER\Documents\ML_Cats_Dogs\train\train'
```

```python
l=os.listdir(path)
```

```python
data=[]

for img in l:
    img_path=os.path.join(path,img)
    arr=cv2.imread(img_path,1)
    arr=cv2.cvtColor(arr,cv2.COLOR_BGR2RGB)
    arr_new=cv2.resize(arr,(100,100))
    img1=img.split('.')
    if(img1[0]=='cat'):
        label=0
    else:
        label=1
    data.append([arr_new, label])
```

```python
data
```

## Shuffling the continuous arrangement data of cats and dogs

```python
import random
random.shuffle(data)
```

```python
x=[]
y=[]
```

```python
for features, label in data:
    x.append(features)
    y.append(label)
```

```python
x=np.array(x)
y=np.array(y)
```

```python
x
```

```python
y
```

```python
x=x/255  #Now all pixelvalues lie from 0 to 255, on dividing by 255 makes the value smaller and the process fast
```

```python
x
```

```python
x.shape   #(no of images, length, breadth, no of channels-red,green and blue channels)
```

## Train model

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential()

model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Conv2D(64, (3,3), activation = 'relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(128, input_shape = x.shape[1:], activation = 'relu'))

model.add(Dense(2, activation = 'softmax'))  #Can use sigmoid function also because they both restrict the value from 0 to 1
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=5, validation_split=0.1)
```
